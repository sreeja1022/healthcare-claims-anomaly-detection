```python
import pandas as pd

df = pd.read_csv('/content/merged_feature_engineered_data.csv')
df.head()
```

```python
# Convert target to numeric
df['PotentialFraud'] = df['PotentialFraud'].map({'Yes': 1, 'No': 0})

# Drop unnecessary columns
X = df.drop(columns=['Provider', 'PotentialFraud'])
y = df['PotentialFraud']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, stratify=y, random_state=42
)
```

```python
from sklearn.ensemble import IsolationForest
from sklearn.metrics import classification_report

iso_model = IsolationForest(contamination=0.1, random_state=42)
y_pred_iso = iso_model.fit_predict(X)

# Convert -1 to fraud (1), 1 to non-fraud (0)
y_pred_iso = [1 if val == -1 else 0 for val in y_pred_iso]

# Evaluate
print("📍 Isolation Forest Report:")
print(classification_report(y, y_pred_iso))
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(random_state=42)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)

print("📍 Random Forest Classification Report:")
print(classification_report(y_test, y_pred_rf))
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

cm = confusion_matrix(y_test, y_pred_rf)
ConfusionMatrixDisplay(cm).plot()
```
